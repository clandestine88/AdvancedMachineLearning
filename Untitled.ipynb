```python
import numpy as np
```

```python
A = np.matrix(([[-2, 1, 0] , [1,-2,1], [0, 1, -2]]))
```

```python
A
```

```python
from scipy import linalg as LA
```

```python
e_vals, e_vecs = LA.eig(A)
```

## Eigenvalues

```python
e_vals
```

## Eigenvectors

```python
e_vecs
```

```python
from numpy.linalg import inv
```

```python
A_inv = inv(A)
```

```python
A_inv
```

```python
A_Ainv = np.matmul(A, A_inv) 
```

```python
A_Ainv
```

```python
Ainv_A = np.matmul(A_inv, A)
```

```python
Ainv_A
```

## As can be seen above, the A*A_inv and A_inv * A both produces very close to identity matrices
